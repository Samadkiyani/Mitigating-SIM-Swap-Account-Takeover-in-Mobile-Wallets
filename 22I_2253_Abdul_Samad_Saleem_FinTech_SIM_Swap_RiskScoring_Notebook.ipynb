
## Problem Statement (SIM‑Swap–Driven Account Takeover in Mobile Wallets)

Mobile wallet and banking users remain vulnerable to **SIM‑swap–driven account takeover (ATO)** when One‑Time Passwords (OTPs) or SMS‑based step‑up codes are used as the primary possession factor. Attackers socially‑engineer mobile carriers to port a victim’s number, then intercept OTPs to reset credentials or authorize transfers. The risk is amplified in **real‑time payment** environments where reversals are difficult. Traditional countermeasures—static device checks, SMS OTP, and post‑event investigations—provide limited protection against modern ATO tactics (reverse‑proxy phishing, MFA fatigue, and automated SIM‑swap campaigns). We need a **phishing‑resistant, risk‑adaptive** approach that (1) reduces reliance on SMS, (2) detects **high‑risk telco and behavioral signals** *before* funds move, and (3) maintains user experience for legitimate customers. This project designs and demonstrates a hybrid solution that combines **FIDO2/WebAuthn**, **device binding & certificate pinning**, and **risk‑based analytics** that fuse **recent SIM‑change indicators** with **behavioral biometrics** to block or step‑up risky transactions automatically.



## Proposed Solution (Phishing‑Resistant MFA + Risk‑Based Analytics + Device Binding)

**Goal.** Prevent SIM‑swap–assisted ATO without degrading good‑user experience. We combine (A) **phishing‑resistant MFA** (FIDO2/WebAuthn), (B) **device binding & pinned TLS** to secure the app‑to‑API channel, and (C) **risk‑based analytics (RBA)** that evaluate telco and behavioral signals in real time.

**A. Phishing‑Resistant MFA (replace SMS OTP).** Per CISA and NIST SP 800‑63B, FIDO2/WebAuthn platform authenticators on iOS/Android are resistant to phishing and **independent of phone numbers**. We enroll users’ devices during KYC or first login, store public keys server‑side, and require WebAuthn for high‑risk actions (password reset, new device login, beneficiary addition, and >PKR X transfers). Roaming authenticators (USB/NFC) serve as recovery.

**B. Device Binding + Certificate Pinning.** The mobile app embeds certificate pinning to prevent MitM, and a device‑bound token (attested key where supported, plus hardware‑backed key storage) ties the session to the physical handset. This thwarts reverse‑proxy OTP theft and session hijacking.

**C. Risk‑Based Analytics Engine.** A small feature set is computed per login or transfer:
- **Telco/Account Signals:** hours since last SIM change (from carrier or third‑party signal providers), number‑to‑SIM mismatch, recent number re‑provision, failed delivery of push notifications.
- **Device/Posture:** first‑seen device, jailbreak/root flags, emulator/automation likelihood, geovelocity, IP reputation.
- **Behavioral Biometrics:** short‑window touch & keystroke rhythms (dwell/flight variance), motion/hold stability, swipe dynamics.
- **Transaction Context:** amount, new payee, out‑of‑profile time/location.

A lightweight classifier (e.g., gradient boosting) outputs a **risk score**. Policy then enforces:
- **Low risk:** silent allow with WebAuthn.
- **Medium risk:** step‑up with **roaming FIDO key** or branch verification.
- **High risk:** *block & alert*, queue for fraud review, and send out‑of‑band notice to **registered** contacts (not SMS to the ported number).

**Privacy & UX.** Behavioral features use short‑term, on‑device summaries (e.g., variances) to avoid storing raw biometrics. False‑positive budgets are controlled by tuning the risk threshold per segment. Enrollment supports **account recovery** paths that do not depend on the potentially hijacked phone number.

**Detection & Response.** High‑risk events generate structured telemetry for SOC/SIEM with playbooks: freeze outgoing RTP/IBFT/RTGS, verify with independent contact, and submit **SIM‑swap abuse reports** to carriers. Dashboards track attack conversion rates and step‑up success.

**Benefits.** (1) Removes SMS OTP from critical flows; (2) detects ATO correlated with **recent SIM change** and **behavioral drift**; (3) preserves smooth UX for the majority via silent allow; (4) aligns with **FFIEC/NIST** guidance and **fast‑payment** risk principles.


## Demonstration (Python Code)

```python

import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve, average_precision_score, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

np.random.seed(123)
N = 12000
hours_since_sim_change = np.random.exponential(scale=600, size=N)
device_first_seen_days = np.random.exponential(scale=220, size=N)
geovelocity_kmph = np.abs(np.random.normal(loc=20, scale=18, size=N))
typing_dwell_var = np.abs(np.random.normal(loc=0.10, scale=0.05, size=N))
swipe_speed_var = np.abs(np.random.normal(loc=0.15, scale=0.06, size=N))
txn_amount_pk = np.abs(np.random.lognormal(mean=2.0, sigma=0.8, size=N))

logit = (
    -4.0
    - 0.0035 * hours_since_sim_change
    - 0.0015 * device_first_seen_days
    + 0.03   * geovelocity_kmph
    + 10.0   * typing_dwell_var
    + 8.5    * swipe_speed_var
    + 0.10   * np.log1p(txn_amount_pk)
)
p_attack = 1 / (1 + np.exp(-logit))
y = (np.random.rand(N) < p_attack).astype(int)

X = pd.DataFrame({
    "hours_since_sim_change": hours_since_sim_change,
    "device_first_seen_days": device_first_seen_days,
    "geovelocity_kmph": geovelocity_kmph,
    "typing_dwell_var": typing_dwell_var,
    "swipe_speed_var": swipe_speed_var,
    "txn_amount_pk": txn_amount_pk,
})

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, stratify=y, random_state=999)

baseline_features = ["hours_since_sim_change", "device_first_seen_days"]
scaler = StandardScaler()
Xb_tr = scaler.fit_transform(X_train[baseline_features])
Xb_te = scaler.transform(X_test[baseline_features])
baseline = LogisticRegression(max_iter=300)
baseline.fit(Xb_tr, y_train)
p_baseline = baseline.predict_proba(Xb_te)[:,1]

proposed = GradientBoostingClassifier(random_state=999)
proposed.fit(X_train, y_train)
p_proposed = proposed.predict_proba(X_test)[:,1]

auc_base = roc_auc_score(y_test, p_baseline)
auc_prop = roc_auc_score(y_test, p_proposed)
ap_base  = average_precision_score(y_test, p_baseline)
ap_prop  = average_precision_score(y_test, p_proposed)

print(f"AUC (Baseline): {auc_base:.3f}")
print(f"AUC (Proposed): {auc_prop:.3f}")
print(f"Average Precision (Baseline): {ap_base:.3f}")
print(f"Average Precision (Proposed): {ap_prop:.3f}")

fpr_b, tpr_b, _ = roc_curve(y_test, p_baseline)
fpr_p, tpr_p, _ = roc_curve(y_test, p_proposed)

plt.figure()
plt.plot(fpr_b, tpr_b, label=f"Baseline ROC (AUC={auc_base:.3f})")
plt.plot(fpr_p, tpr_p, label=f"Proposed ROC (AUC={auc_prop:.3f})")
plt.plot([0,1],[0,1],'--', label="Chance")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Comparison: Baseline vs Proposed")
plt.legend(loc="lower right")
plt.show()

def tpr_at_fpr(fpr, tpr, target=0.01):
    idx = np.argmin(np.abs(fpr - target))
    return tpr[idx]

tpr_b_at_1 = tpr_at_fpr(fpr_b, tpr_b, 0.01)
tpr_p_at_1 = tpr_at_fpr(fpr_p, tpr_p, 0.01)
print(f"TPR at 1% FPR - Baseline: {tpr_b_at_1:.3f}, Proposed: {tpr_p_at_1:.3f}")
```


## Results & Comparison

On the synthetic evaluation, the **proposed fused risk model** substantially outperforms the **baseline (OTP‑only + coarse checks)**:

- **ROC‑AUC** and **Average Precision** are markedly higher for the proposed model.
- At a fixed **1% false positive rate**, the proposed model achieves a higher **true positive rate**, enabling earlier blocks/step‑ups with minimal friction for legitimate users.
- This supports adopting **phishing‑resistant MFA** plus **risk‑based analytics** that include **recent SIM‑change signals** and **behavioral biometrics** over legacy SMS OTP flows.
